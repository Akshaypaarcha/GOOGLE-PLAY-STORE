```python
data['Installs'] = data['Installs'].str.split('+').str[0]
data['Installs'].replace(',','',regex=True, inplace=True)
data['Installs'] = data['Installs'].astype(np.int64)
```

```python
data['Installs'].unique()
```

```python
data['Size'].unique()
```

```python
data['Content Rating'].unique()
```

```python
data['Size'] = data['Size'].apply(lambda x:str(x).replace('M','')  if 'M' in str(x) else x)
```

```python
data['Size'] = data['Size'].apply(lambda x:str(x).replace(',','.') if ',' in str(x) else x)
```

```python
data['Size'] = data['Size'].apply(lambda x:float(str(x).replace('k',''))/1000 if 'k' in str(x) else x)
```

```python
data['Size'] = data['Size'].apply(lambda x: str(x).replace('Varies with device','0') if 'Varies with device' in str(x) else(x))
```

```python
data['Size'] = data['Size'].apply(lambda x:float(str(x).replace('G',''))*1000 if 'G' in str(x) else x)
```

```python
data['Size'] = data['Size'].apply(lambda x:float(x))
```
